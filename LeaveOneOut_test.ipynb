```python
def cap_outliers(df, col_name, *args):
    if len(args) > 0:
        upper, lower = args
        print (upper, lower)

    else:
        Q1 = np.percentile(df[col_name], 25, interpolation = 'midpoint')
        Q3 = np.percentile(df[col_name], 75, interpolation = 'midpoint')
        IQR = Q3 - Q1

        # Upper bound
        upper = Q3+1.5*IQR
        # Lower bound
        lower = Q1-1.5*IQR

    ''' Capping the Outliers '''
    df.loc[df[col_name] >= upper, col_name] = upper
    df.loc[df[col_name] <= lower, col_name] = lower
    return df, upper, lower
for col in columns:
    df_train, upper, lower = cap_outliers(df_train, col)
    df_test, upper, lower = cap_outliers(df_test, col, upper, lower)
```

```python
from sklearn.preprocessing import StandardScaler
import pickle    
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
pickle.dump(scaler, open('scaler.pkl','wb'))
```

```python
from time import time
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

names = [ 
    "RandomForest",
    "NeuralNetMLP",
    "XGB"
]
classifiers = [
    RandomForestClassifier(),
    MLPClassifier(max_iter=2000),
    XGBClassifier()
]

max_scores_mean=0
model_name=""
speed={}
accuracy={}
balanced_accuracy_scores={}
accuracy_on_test_set={}
precision_scores={}
recall_scores={}
f1_scores={}
for name, clf in zip(names, classifiers):
    print(name)
    clf.fit(X_train, y_train)
    start = time()
    skf = StratifiedKFold(10, shuffle=True, random_state=1)
    scores=cross_val_score(clf,X_train,y_train,cv=skf, scoring="accuracy")
    speed[name] = np.round(time() - start, 3)
    accuracy[name] = np.mean(scores).round(3)
    
    print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
    model_name = name
    filename = name+"_model.sav"
    pickle.dump(clf, open(filename, 'wb'))
    print()
    print(model_name)
    print("cross_val_score:",scores.mean())
#     print("real value:", y_test.values.tolist())
#     print("predicted value:",clf.predict(X_test).tolist())
    print("score:",clf.score(X_test, y_test))
    print("balanced_accuracy_score:", balanced_accuracy_score(y_test,clf.predict(X_test)))
    print("precision_score:", precision_score(y_test,clf.predict(X_test), average='weighted'))
    print("recall_score:", recall_score(y_test,clf.predict(X_test), average='weighted'))
    print("f1_score:", f1_score(y_test,clf.predict(X_test),  average='weighted'))
    
    balanced_accuracy_scores[name] = balanced_accuracy_score( y_test, clf.predict(X_test))
    accuracy_on_test_set[name] = clf.score(X_test, y_test)
    precision_scores[name] = precision_score(y_test,clf.predict(X_test), average='weighted')
    recall_scores[name] = recall_score(y_test,clf.predict(X_test), average='weighted')
    f1_scores[name] = f1_score(y_test,clf.predict(X_test), average='weighted')
print("Cross validation accuracy:")
print({k:v for k,v in sorted(accuracy.items(), key= lambda i: i[1], reverse=True)})
print()
print("Accuracy on test set:")
print({k:v for k,v in sorted(accuracy_on_test_set.items(), key= lambda i: i[1], reverse=True)})
print()
print("balanced_accuracy_scores on test set:")
print({k:v for k,v in sorted(balanced_accuracy_scores.items(), key= lambda i: i[1], reverse=True)})
print()
print("precision_score on test set:")
print({k:v for k,v in sorted(precision_scores.items(), key= lambda i: i[1], reverse=True)})
print()
print("recall_score on test set:")
print({k:v for k,v in sorted(recall_scores.items(), key= lambda i: i[1], reverse=True)})
print()
print("f1_score on test set:")
print({k:v for k,v in sorted(f1_scores.items(), key= lambda i: i[1], reverse=True)})
print("Speed:")
print({k:v for k,v in sorted(speed.items(), key= lambda i: i[1], reverse=False)})
```
